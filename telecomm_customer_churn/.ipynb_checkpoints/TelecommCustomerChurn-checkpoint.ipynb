```python
import numpy as np
import pandas as pd
from IPython.display import display
data = pd.ExcelFile("WA_Fn-UseC_-Telco-Customer-Churn.xlsx").parse()
churned = data['Churn']
features = data.drop({'customerID','Churn'}, axis = 1)
#display(features)

churned.replace({'No':0,'Yes':1},inplace=True)
#display(churned)
features['gender'].replace({'Female':0,'Male':1},inplace=True)
features['Partner'].replace({'No':0,'Yes':1},inplace=True)
features['Dependents'].replace({'No':0,'Yes':1},inplace=True)
features['PhoneService'].replace({'No':0,'Yes':1},inplace=True)
features['MultipleLines'].replace({'No':0,'Yes':1, 'No phone service':2},inplace=True)
features['InternetService'].replace({'No':0,'DSL':1, 'Fiber optic':2},inplace=True)
features['OnlineSecurity'].replace({'No':0,'Yes':1, 'No internet service':2},inplace=True)
features['OnlineBackup'].replace({'No':0,'Yes':1, 'No internet service':2},inplace=True)
features['DeviceProtection'].replace({'No':0,'Yes':1, 'No internet service':2},inplace=True)
features['TechSupport'].replace({'No':0,'Yes':1, 'No internet service':2},inplace=True)
features['StreamingTV'].replace({'No':0,'Yes':1, 'No internet service':2},inplace=True)
features['StreamingMovies'].replace({'No':0,'Yes':1, 'No internet service':2},inplace=True)
features['Contract'].replace({'Month-to-month':0,'One year':1, 'Two year':2},inplace=True)
features['PaperlessBilling'].replace({'No':0,'Yes':1},inplace=True)
features['PaymentMethod'].replace({'Bank transfer (automatic)':0,'Credit card (automatic)':1, 
                                   'Electronic check':2, 'Mailed check':3},inplace=True)
#features['TotalCharges'].convert_objects(convert_numeric=True)  float64
features['TotalCharges'] = pd.to_numeric(features['TotalCharges'], errors='coerce') .fillna(0).astype(np.int64)
display(features.iloc[488])
```

```python
from sklearn.model_selection import train_test_split

print ("Total Data ", len(features))
X_train, X_test, y_train, y_test = train_test_split(features, churned, test_size=0.20, random_state=42)
print ("train ",len(X_train))
print ("test ",len(X_test))
print ("Training and testing split was successful.")
                   
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import r2_score
from sklearn.metrics import make_scorer

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    return r2_score(y_true, y_predict)

def fit_model(X, y):
    """ Performs grid search over the 'max_depth' parameter for a 
        decision tree regressor trained on the input data [X, y]. """
    
    # Create cross-validation sets from the training data
    # sklearn version 0.18: ShuffleSplit(n_splits=10, test_size=0.1, train_size=None, random_state=None)
    # sklearn versiin 0.17: ShuffleSplit(n, n_iter=10, test_size=0.1, train_size=None, random_state=None)
    cv_sets = ShuffleSplit(X.shape[0], test_size = 0.20, random_state = 0)

    # : Create a decision tree classifier object
    classifier = DecisionTreeClassifier(random_state=0)

    # : Create a dictionary for the parameter 'max_depth' with a range from 1 to 10
    params = {'max_depth':np.arange(1, 11, dtype=np.int).tolist()}

    # : Transform 'performance_metric' into a scoring function using 'make_scorer' 
    scoring_fnc = make_scorer(performance_metric)

    # : Create the grid search cv object --> GridSearchCV()
    # Make sure to include the right parameters in the object:
    # (estimator, param_grid, scoring, cv) which have values 'regressor', 'params', 'scoring_fnc', and 'cv_sets' respectively.
    grid = GridSearchCV(classifier, params, scoring_fnc, cv = cv_sets)

    # Fit the grid search object to the data to compute the optimal model
    grid = grid.fit(X, y)

    # Return the optimal model after fitting the data
    return grid.best_estimator_
```

```python
reg = fit_model(X_train, y_train)
print ("Parameter 'max_depth' is {} for the optimal model.".format(reg.get_params()['max_depth']))
```
